

#### ✅ **우선순위 2: 슬리피지 모델링**


```python
# 체결가격 = 기대가격 + 슬리피지
def calculate_slippage(expected_price, volume_ratio):
    """
    volume_ratio: 일일 거래량 대비 주문 비율
    """
    if volume_ratio < 0.01:  # 1% 미만
        slippage = 0.0001  # 0.01%
    elif volume_ratio < 0.05:  # 5% 미만
        slippage = 0.0005  # 0.05%
    else:
        slippage = 0.001  # 0.1%
    
    return expected_price * (1 + slippage)
```

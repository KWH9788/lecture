```python

# 1) 라이브러리 임포트 & 환경 설정
import os, gc, warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn-v0_8')
sns.set_palette('Set2')
pd.set_option('display.max_columns', 100)

# 파일 경로 (업로드된 경로 유지)
DATA_PATH = '../Data_Analyze_data/iot_telemetry_data.csv'

# 유틸 함수
def mem_usage(df):
    return df.memory_usage(index=True, deep=True).sum() / (1024**2)

def info_compact(df, head=3):
    display(df.head(head))
    display(df.tail(head))
    print(f'행/열: {df.shape[0]:,} x {df.shape[1]}')
    print(f'메모리 사용량: {mem_usage(df):.2f} MB')
    print('\n컬럼 타입:')
    print(df.dtypes)

RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)

import platform
if platform.system() == "Darwin":
    plt.rc("font", family="AppleGothic")
else:
    plt.rc("font", family="Malgun Gothic")
```

```python

# 2) 데이터 로딩
# dtype 유추 + ts 파싱
raw = pd.read_csv(
    DATA_PATH,
    low_memory=False
)

# 컬럼 표준 이름 가정
# 예상 컬럼: ts, device, co, lpg, smoke, temp, humidity, light, motion
raw.columns = [c.strip().lower() for c in raw.columns]

raw.replace(['00:0f:00:70:91:0a', '1c:bf:ce:15:ec:4d', 'b8:27:eb:bf:9d:51'], [1, 2, 3], inplace=True)

# ts 파싱 (UTC 가정)
if 'ts' in raw.columns:
    raw['ts'] = pd.to_datetime(raw['ts'], unit='s', utc=True)
else:
    raise ValueError('타임스탬프(ts) 컬럼을 찾을 수 없습니다.')

# motion이 정수/불리언이 아니면 정수로 캐스팅 시도
if 'motion' in raw.columns:
    if not pd.api.types.is_integer_dtype(raw['motion']):
        raw['motion'] = pd.to_numeric(raw['motion'], errors='coerce').fillna(0).astype(int)

info_compact(raw)
raw.head(10)
```

```python

# 3) 기본 유효성 점검
# 누락치 카운트
na_cnt = raw.isna().sum().sort_values(ascending=False)
print('결측치 요약:')
display(na_cnt.to_frame('n_missing'))

# 시간 정렬 및 중복/역행 확인
df = raw.sort_values('ts').reset_index(drop=True)
is_monotonic = df['ts'].is_monotonic_increasing
dup_ts = df['ts'].duplicated().sum()

print(f'시간 오름차순 정렬 여부(정렬 후): {is_monotonic}')
print(f'중복 타임스탬프 개수: {dup_ts:,}')

# 타임 범위
t_min, t_max = df['ts'].min(), df['ts'].max()
print(f'시간 범위(UTC): {t_min} ~ {t_max} (총 {(t_max - t_min)})')

# 디바이스 현황
if 'device' in df.columns:
    n_devices = df['device'].nunique()
    print(f'디바이스 수: {n_devices}')
    dev_counts = df['device'].value_counts().rename('rows').to_frame()
    display(dev_counts)
else:
    print('device 컬럼이 없습니다. 단일 소스 데이터로 간주합니다.')
```

```python

# 4) 디바이스별 샘플링 간격 추정(초)
def sampling_stats(g):
    dt = g['ts'].diff().dt.total_seconds().dropna()
    if len(dt)==0:
        return pd.Series({'n': len(g), 'median_s': np.nan, 'mean_s': np.nan, 'p10_s': np.nan, 'p90_s': np.nan})
    return pd.Series({
        'n': len(g),
        'median_s': float(np.median(dt)),
        'mean_s': float(np.mean(dt)),
        'p10_s': float(np.percentile(dt, 10)),
        'p90_s': float(np.percentile(dt, 90)),
    })

if 'device' in df.columns:
    samp = df.groupby('device', sort=False).apply(sampling_stats).reset_index()
else:
    samp = sampling_stats(df).to_frame().T
display(samp)
```

```python

# 5) 결측 히트맵
plt.figure(figsize=(10, 3))
sns.heatmap(df.isna(), cbar=False)
plt.title('Missingness heatmap')
plt.tight_layout()
plt.show()
```

```python

# 6) 수치형 컬럼 요약 통계
num_cols = [c for c in df.columns if c not in ['ts','device'] and pd.api.types.is_numeric_dtype(df[c])]
cat_cols = [c for c in df.columns if c not in ['ts'] and not pd.api.types.is_numeric_dtype(df[c])]

desc = df[num_cols].describe(percentiles=[.01,.05,.25,.5,.75,.95,.99]).T
desc['missing'] = df[num_cols].isna().sum()
display(desc)
print('범주형 후보 컬럼:', cat_cols)
```

```python

# 7) 변수 분포(히스토그램/커널밀도)
n = len(num_cols)
ncols = 3
nrows = (n + ncols - 1)//ncols if n>0 else 1
plt.figure(figsize=(ncols*5, max(1,nrows)*3.5))
for i, c in enumerate(num_cols, 1):
    ax = plt.subplot(nrows, ncols, i)
    sns.histplot(df[c], kde=True, bins=50, ax=ax)
    ax.set_title(c)
plt.tight_layout()
plt.show()
```

```python
# 8) 시계열 개요(부분 구간 플롯)
# 폭주 방지를 위해 첫 6시간 구간만 샘플링
t0 = df['ts'].min()
t1 = t0 + pd.Timedelta(hours=6)
df_sample = df[(df['ts']>=t0)&(df['ts']<=t1)].copy()

plot_cols = [c for c in num_cols if c not in ['motion']]  # motion은 별도
if len(df_sample) > 0 and len(plot_cols)>0:
    n = len(plot_cols)
    ncols = 2
    nrows = (n + ncols - 1)//ncols
    fig, axes = plt.subplots(nrows, ncols, figsize=(ncols*6, nrows*2.8), sharex=True)
    axes = axes.ravel() if n>1 else [axes]
    for i, c in enumerate(plot_cols):
        ax = axes[i]
        if 'device' in df_sample.columns:
            for d, sub in df_sample.groupby('device'):
                ax.plot(sub['ts'], sub[c], label=str(d), alpha=0.8)
            ax.legend(fontsize=8, ncol=2)
        else:
            ax.plot(df_sample['ts'], df_sample[c], color='C0')
        ax.set_title(c)
    # 남는 축 비우기
    for j in range(i+1, len(axes)):
        axes[j].axis('off')
    plt.tight_layout()
    plt.show()

# motion 비율(샘플 구간)
if 'motion' in df_sample.columns:
    motion_rate = df_sample['motion'].mean()
    print(f'샘플 구간 motion 비율: {motion_rate:.3f}')
```

```python

# 9) 상관 행렬(피어슨)
if len(num_cols) >= 2:
    corr = df[num_cols].corr(method='pearson')
    plt.figure(figsize=(8,6))
    sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', vmin=-1, vmax=1)
    plt.title('Correlation matrix (Pearson)')
    plt.tight_layout()
    plt.show()
```

```python

# 10) KST 변환 및 일중/요일 패턴
df_kst = df.copy()
df_kst['ts_kst'] = df_kst['ts'].dt.tz_convert('Asia/Seoul')
df_kst['date'] = df_kst['ts_kst'].dt.date
df_kst['hour'] = df_kst['ts_kst'].dt.hour
df_kst['dow'] = df_kst['ts_kst'].dt.dayofweek  # 0=월

# 시간대별 평균
agg_cols = [c for c in num_cols if c != 'motion']
if len(agg_cols)>0:
    hourly = df_kst.groupby('hour')[agg_cols].mean()
    plt.figure(figsize=(10,4))
    hourly.plot(ax=plt.gca())
    plt.title('일중(hour) 평균 (KST)')
    plt.xlabel('hour (KST)')
    plt.tight_layout()
    plt.show()

# motion 일중 패턴
if 'motion' in df_kst.columns:
    hourly_m = df_kst.groupby('hour')['motion'].mean()
    plt.figure(figsize=(8,3))
    hourly_m.plot(marker='o')
    plt.title('일중(hour) 평균 motion 비율 (KST)')
    plt.xlabel('hour (KST)')
    plt.ylabel('motion rate')
    plt.tight_layout()
    plt.show()

# 디바이스별 일중 패턴 (상위 4변수 예시)
if 'device' in df_kst.columns and len(agg_cols)>0:
    for c in agg_cols[:4]:
        plt.figure(figsize=(10,4))
        pivot = df_kst.groupby(['device','hour'])[c].mean().unstack(0)
        pivot.plot(title=f'일중 패턴 (KST) - {c}', ax=plt.gca())
        plt.xlabel('hour')
        plt.tight_layout()
        plt.show()
```

```python

# 11) 리샘플(1분/1시간) 요약
df_min = df.set_index('ts').sort_index()
rule = '1min'
agg_map = {c:'mean' for c in num_cols}
min_agg = df_min.resample(rule).agg(agg_map)
display(min_agg.head())

plt.figure(figsize=(10,4))
for c in [col for col in ['temp','humidity','light'] if col in min_agg.columns]:
    plt.plot(min_agg.index, min_agg[c], label=c, alpha=0.9)
plt.legend(); plt.title('1분 평균 (예: temp/humidity/light)'); plt.tight_layout(); plt.show()

# 시간별
hr_agg = df_min.resample('1H').agg(agg_map)
display(hr_agg.describe().T)
```

```python

# 12) 간단 스파이크 탐지(롤링 Z-score)
target_cols = [c for c in ['co','lpg','smoke','light'] if c in df.columns]
if target_cols:
    win = 60  # 관측치 개수 기준 윈도우
    thr = 3.0
    sdf = df.set_index('ts').sort_index().copy()
    for c in target_cols:
        s = sdf[c].astype(float)
        mu = s.rolling(win, min_periods=10).mean()
        sd = s.rolling(win, min_periods=10).std()
        z = (s - mu) / (sd + 1e-9)
        spikes = (z.abs() > thr)
        n_spikes = int(spikes.sum())
        print(f'{c}: z>|{thr}| 스파이크 개수 = {n_spikes}')
        # 샘플 플롯(앞 3시간)
        t0 = sdf.index.min()
        t1 = t0 + pd.Timedelta(hours=3)
        sub = sdf.loc[(sdf.index>=t0)&(sdf.index<=t1), [c]].copy()
        sub['z'] = z.loc[sub.index]
        sub['spike'] = spikes.loc[sub.index]
        fig, ax1 = plt.subplots(figsize=(10,3))
        ax1.plot(sub.index, sub[c], label=c, color='C0')
        ax1_t = ax1.twinx()
        ax1_t.plot(sub.index, sub['z'], color='C1', alpha=0.6, label='z-score')
        ax1_t.axhline(thr, color='C1', ls='--', alpha=0.5)
        ax1_t.axhline(-thr, color='C1', ls='--', alpha=0.5)
        ax1.scatter(sub.index[sub['spike']], sub[c][sub['spike']], color='red', s=10, label='spike')
        ax1.set_title(f'{c} - 스파이크(샘플 3시간)')
        fig.legend(loc='upper right')
        plt.tight_layout(); plt.show()
```

```python

# 13) 산점도/박스 그림 예시
if 'motion' in df.columns and 'light' in df.columns:
    # 샘플링으로 점 수 축소
    sample = df.sample(min(len(df), 30000), random_state=RANDOM_STATE)
    plt.figure(figsize=(5,4))
    sns.boxenplot(x='motion', y='light', data=sample, palette='Set2')
    plt.title('Light by Motion')
    plt.tight_layout(); plt.show()

if 'temp' in df.columns and 'humidity' in df.columns:
    sample = df.sample(min(len(df), 30000), random_state=RANDOM_STATE)
    plt.figure(figsize=(5,4))
    sns.scatterplot(x='temp', y='humidity', hue=sample['motion'] if 'motion' in sample.columns else None, s=8, alpha=0.5, data=sample)
    plt.title('Temp vs Humidity (hue=motion)')
    plt.tight_layout(); plt.show()
```

```python

# 14) 디바이스 드리프트 점검(일평균 추이)
if 'device' in df_kst.columns:
    daily = df_kst.groupby(['device','date']).agg({c:'mean' for c in [col for col in ['co','lpg','smoke','temp','humidity','light'] if col in df_kst.columns]})
    if not daily.empty:
        display(daily.head())
        for c in daily.columns[:4]:
            pivot = daily[c].unstack(0)
            pivot.plot(marker='o', figsize=(8,3), title=f'일평균 추이 - {c}')
            plt.tight_layout(); plt.show()
```

```python

# 15) 간단 요약 출력
summary = {}
summary['rows'] = int(len(df))
summary['cols'] = int(df.shape[1])
summary['t_min_utc'] = str(df['ts'].min())
summary['t_max_utc'] = str(df['ts'].max())
summary['duration'] = str(df['ts'].max() - df['ts'].min())
summary['n_devices'] = int(df['device'].nunique()) if 'device' in df.columns else 1
summary['numeric_cols'] = num_cols

print('요약:')
for k,v in summary.items():
    print(f'- {k}: {v}')
```
